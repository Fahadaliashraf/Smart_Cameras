```python
import cv2
from ultralytics import YOLO
import pyttsx3
import time
```

```python
model = YOLO("yolov8n.pt")  # Automatically download ho jayega
```

```python
engine = pyttsx3.init()

def speak(text):
    engine.say(text)
    engine.runAndWait()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print(" Webcam nahi chal rahi. Check if it's busy or blocked.")
else:
    print(" Webcam chal gayi!")

# Try reading one frame
ret, frame = cap.read()

if not ret:
    print(" Frame nahi mil rahi. Ho sakta hai koi permission ya hardware issue ho.")
else:
    print(" Frame mil gayi! Shape:", frame.shape)
```
